# Exploratory Data Analysis of Suicides in India


Exploratory data analaysis is performed on the dataset Sucides in India obtained from kaggle

Dataset contains data from 2001-2012


Dataset:https://www.kaggle.com/rajanand/suicides-in-india

```python
# importing required libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set()
%matplotlib inline
```

```python
# reading the dataset using pandas library
df = pd.read_csv("Suicides in India.csv")
```

```python
#displaying the first 5 rows of the dataset
df.head()
```

```python
#displaying the last 5 rows of the dataset
df.tail()
```

```python
df.State.value_counts()
```

```python
df.drop(df[(df.State=='Total (States)')|(df.State=='Total (All India)')|(df.State=='Total (Uts)')].index,inplace=True)
```

```python
#plotting no of suicides based on gender
Gender=df.groupby('Gender').sum()['Total'].plot(kind='bar',title='Count of suicides Gender wise',figsize=(5,6))
Gender.set_xlabel("Gender")
Gender.set_ylabel("No of suicides")
```

```python
#plotting n
states = df.groupby('State').sum()['Total'].sort_values(ascending=True).plot(kind='bar', figsize= (15,8), title ="State wise suicides frequency")
states.set_xlabel('\nState')
states.set_ylabel('Counts\n')
```

```python
male=df[df['Gender']=='Male']
female=df[df['Gender']=='Female']
```

```python
causes=df[df['Type_code']=='Causes']
education=df[df['Type_code']=='Education_Status']
means=df[df['Type_code']=='Means_adopted']
profile=df[df['Type_code']=='Professional_Profile']
social=df[df['Type_code']=='Social_Status']
```

```python
#Age Suicide Frequency
data_age=df[df['Age_group']!='0-100+']

age = data_age.groupby('Age_group').sum()['Total'].plot(kind="bar",figsize=(8,5), title ="Age wise suicides frequency")
age.set_xlabel('\nAge Group')
age.set_ylabel('Counts\n')
```

```python
Year=df.groupby('Year').sum()['Total'].plot(kind='line',title='Count of suicides Year Wise')
Year.set_xlabel("year")

Year.set_ylabel("No of suicides")
```

```python
Social = social.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(8,5), title ="No of Suicides vs Social Status")
Social.set_xlabel('\nSocial Status')
Social.set_ylabel('Counts\n')
```

```python
Profile = profile.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(10,6), title ="No of Suicides vs Professional Status")
Profile.set_xlabel('\nProfessional Status')
Profile.set_ylabel('Counts\n')
```

```python
Means = means.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(13,7), title ="No of Suicides vs Means adopted")
Means.set_xlabel('\nMeans Adopted')
Means.set_ylabel('Counts\n')
```

```python
Education = education.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(8,5), title ="No of Suicides vs Education Status")
Education.set_xlabel('\nEducation Status')
Education.set_ylabel('Counts\n')
```

```python
Causes = causes.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(15,8), title ="No of Suicides vs Causes")
Causes.set_xlabel('\nCauses')
Causes.set_ylabel('Counts\n')
```

```python
#Age Suicide Frequency
data_age1=causes[causes['Age_group']=='0-14']


age = data_age1.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(15,8), title ="No of suicides in the age group 0-14")
age.set_xlabel('\nCauses')
age.set_ylabel('Counts\n')
```

```python
data_age2=causes[causes['Age_group']=='15-29']


age = data_age2.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(15,8), title ="No of suicides in the age group 15-29")
age.set_xlabel('\nCauses')
age.set_ylabel('Counts\n')
```

```python
data_age3=causes[causes['Age_group']=='30-44']


age = data_age3.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(15,8), title ="No of suicides in the age group 30-44")
age.set_xlabel('\nCauses')
age.set_ylabel('Counts\n')
```

```python
data_age4=causes[causes['Age_group']=='45-59']


age = data_age4.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(15,8), title ="No of suicides in the age group 45-59")
age.set_xlabel('\nCauses')
age.set_ylabel('Counts\n')
```

```python
data_age5=causes[causes['Age_group']=='60+']


age = data_age5.groupby('Type').sum()['Total'].sort_values(ascending=True).plot(kind="bar",figsize=(15,8), title ="No of suicides in the age group 60+")
age.set_xlabel('\nCauses')
age.set_ylabel('Counts\n')
```

## Conclusion


- Age group between 15-29 has highest nunber of suicides
- As per the dataset Maharashtra, West Bengal, Tamil Nadu, Andhra Pradesh, Karnataka has the highest no of suicides over the years
